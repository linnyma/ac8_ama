### Importación de datos



```python
# Importación de datos

import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Lectura de datos

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Tienda 1
tienda.head()
```

#1. Análisis de facturación



```python
# Ingreso total

# Tienda 1

total_tienda1 = sum(tienda['Precio'])

# Tienda 2

total_tienda2 = sum(tienda2['Precio'])

# Tienda 3

total_tienda3 = sum(tienda3['Precio'])

# Tienda 4

total_tienda4 = sum(tienda4['Precio'])

# Impresión de datos

print(f'Total de ventas por tiendas: \nTienda 1: {total_tienda1:,.2f} \nTienda 2: {total_tienda2:,.2f} \nTienda 3: {total_tienda3:,.2f} \nTienda 4: {total_tienda4:,.2f}')
```

```python
# Calculación de porcentajes sobre el total

venta_total = (total_tienda1 + total_tienda2 + total_tienda3 + total_tienda4)

# Porcentajes

# Tienda 1

porcentaje_ventas1 = ((total_tienda1 * 100) / venta_total)

# Tienda 2

porcentaje_ventas2 = ((total_tienda2 * 100) / venta_total)

# Tienda 3

porcentaje_ventas3 = ((total_tienda3 * 100) / venta_total)

# Tienda 4

porcentaje_ventas4 = ((total_tienda4 * 100) / venta_total)

# Impresión de datos

print(f'Total de porcentaje de ventas por tiendas sobre el total: \nTienda 1: {porcentaje_ventas1:,.2f}% \nTienda 2: {porcentaje_ventas2:,.2f}% \nTienda 3: {porcentaje_ventas3:,.2f}% \nTienda 4: {porcentaje_ventas4:,.2f}%')
```

# 2. Ventas por categoría

```python
# Conteo de total de ventas por categoría por tienda

columna_categorias1 = tienda['Categoría del Producto'].value_counts()
columna_categorias2 = tienda2['Categoría del Producto'].value_counts()
columna_categorias3 = tienda3['Categoría del Producto'].value_counts()
columna_categorias4 = tienda4['Categoría del Producto'].value_counts()

# Creación de un nuevo dataframe con la información de las ventas por categorías

tabla_categorias = pd.concat([columna_categorias1, columna_categorias2, columna_categorias3, columna_categorias4], axis=1)
tabla_categorias.columns = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

tabla_categorias
```

# 3. Calificación promedio de la tienda


```python
# Cálculo de media general de satisfacción por tienda en un rango del 1 al 5:

media_tienda1 = tienda['Calificación'].mean()
media_tienda2 = tienda2['Calificación'].mean()
media_tienda3 = tienda3['Calificación'].mean()
media_tienda4 = tienda4['Calificación'].mean()

print(f'Media de satisfacción del cliente por tiendas: \nTienda 1 - {media_tienda1:,.2f} \nTienda 2 - {media_tienda2:,.2f} \nTienda 3 - {media_tienda3:,.2f} \nTienda 4 - {media_tienda4:,.2f}')
```

# 4. Productos más y menos vendidos

```python
# Creación de DataFrames separados para más y menos vendidos por tienda

mas_vendidos1 = tienda['Producto'].value_counts().head(5).to_frame(name='Ventas')
mas_vendidos1['Tienda'] = 'Tienda 1'
mas_vendidos1['Tipo'] = 'Más vendidos'

menos_vendidos1 = tienda['Producto'].value_counts().tail(5).to_frame(name='Ventas')
menos_vendidos1['Tienda'] = 'Tienda 1'
menos_vendidos1['Tipo'] = 'Menos vendidos'

mas_vendidos2 = tienda2['Producto'].value_counts().head(5).to_frame(name='Ventas')
mas_vendidos2['Tienda'] = 'Tienda 2'
mas_vendidos2['Tipo'] = 'Más vendidos'

menos_vendidos2 = tienda2['Producto'].value_counts().tail(5).to_frame(name='Ventas')
menos_vendidos2['Tienda'] = 'Tienda 2'
menos_vendidos2['Tipo'] = 'Menos vendidos'

mas_vendidos3 = tienda3['Producto'].value_counts().head(5).to_frame(name='Ventas')
mas_vendidos3['Tienda'] = 'Tienda 3'
mas_vendidos3['Tipo'] = 'Más vendidos'

menos_vendidos3 = tienda3['Producto'].value_counts().tail(5).to_frame(name='Ventas')
menos_vendidos3['Tienda'] = 'Tienda 3'
menos_vendidos3['Tipo'] = 'Menos vendidos'

mas_vendidos4 = tienda4['Producto'].value_counts().head(5).to_frame(name='Ventas')
mas_vendidos4['Tienda'] = 'Tienda 4'
mas_vendidos4['Tipo'] = 'Más vendidos'

menos_vendidos4 = tienda4['Producto'].value_counts().tail(5).to_frame(name='Ventas')
menos_vendidos4['Tienda'] = 'Tienda 4'
menos_vendidos4['Tipo'] = 'Menos vendidos'


# Concatenar todo en una sola tabla

tabla_final = pd.concat([
    mas_vendidos1, menos_vendidos1,
    mas_vendidos2, menos_vendidos2,
    mas_vendidos3, menos_vendidos3,
    mas_vendidos4, menos_vendidos4
])

# Renombrar índice
tabla_final.index.name = 'Producto'

# Mostrar resultado
tabla_final
```

# 5. Envío promedio por tienda

```python
# Cálculo de media general de costos de envío por tienda

envio_promedio1 = tienda['Costo de envío'].mean()
envio_promedio2 = tienda['Costo de envío'].mean()
envio_promedio3 = tienda['Costo de envío'].mean()
envio_promedio4 = tienda['Costo de envío'].mean()

print(f'Media de costos de envío por tiendas: \nTienda 1 - {envio_promedio1:,.2f} \nTienda 2 - {envio_promedio2:,.2f} \nTienda 3 - {envio_promedio3:,.2f} \nTienda 4 - {envio_promedio4:,.2f}')
```

```python
# Importar librería

import matplotlib.pyplot as plt
```

```python
# Gráfico de análisis de facturación

tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

totales = [total_tienda1, total_tienda2, total_tienda3, total_tienda4]

porcentaje_general = [porcentaje_ventas1, porcentaje_ventas2, porcentaje_ventas3, porcentaje_ventas4]

fig, ax1 = plt.subplots(figsize=(8, 5))

# Barras en eje venta

barras = ax1.bar(tiendas, totales, color='#B2DDEB', edgecolor='black', label='Ventas totales ($)')
ax1.set_xlabel('Tiendas')
ax1.set_ylabel('Facturación ($)', color='cornflowerblue')
ax1.tick_params(axis='y', labelcolor='cornflowerblue')
ax1.grid(axis='y', linestyle='--', alpha=0.6)

# Eje para porcentajes

ax2 = ax1.twinx()
ax2.plot(tiendas, porcentaje_general, color='#0C5061', marker='o', linestyle='--', linewidth=2, label='Porcentaje sobre el total (%)')
ax2.set_ylabel('Porcentaje (%)', color='tomato')
ax2.tick_params(axis='y', labelcolor='tomato')

# Personalización

plt.title('Análisis de facturación por tienda')
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
plt.legend(lines + lines2, labels + labels2, loc='upper left')

plt.show()

```
